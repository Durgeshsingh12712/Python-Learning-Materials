```python
x = [1, 2, 3, 4, 5]
y = [5, 1, 7, 8, 2]
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = 'red')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = 'red', marker = 'o')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', linewidth = 3, markersize = '7')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '--', linewidth = 3, markersize = '7')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]

plt.plot(x, y1, color = 'red', marker = 'o', linestyle = '--', linewidth = 3, markersize = '6', label = 'line y1')
plt.plot(x, y2, color = 'blue', marker = 'o', linestyle = '--', linewidth = 3, markersize = '7', label = 'line y2')

plt.xlabel("x data axis")
plt.ylabel("y data axis")

plt.title("Multiple line in one plot")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))

plt.show
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x+1))

plt.show
```

```python
data1 = np.random.normal(0, 1, 5000)
data2 = np.random.normal(3, 2, 5000)
```

```python
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True)
plt.title("Pie Chart")
plt.show()
```
